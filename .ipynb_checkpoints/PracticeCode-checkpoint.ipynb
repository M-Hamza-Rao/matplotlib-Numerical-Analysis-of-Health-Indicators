### Scatterplot

```python
plt.figure(figsize=(4, 3))
plt.scatter(df['Sales'], df['Profit'], color='blue', alpha=0.5, s=20)

plt.xticks(np.arange(100, 801, 50), fontsize=7)
plt.title('Sales vs. Profit Scatter Plot', fontsize=7)
plt.xlabel('Sales', fontsize=6)
plt.ylabel('Profit', fontsize=6)

# Show grid
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```

### Bar Chart

```python
sales_by_region = df.groupby('Region')['Sales'].sum()


plt.figure(figsize=(5, 3))
sales_by_region.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen', 'pink'])


plt.title('Total Sales by Region (2023)', fontsize=12)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.yticks(np.arange(0, 2000, 200), fontsize=7)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### BoxPlot

```python
plt.figure(figsize=(6, 4))
box = df.boxplot(column='Sales', by='Region', patch_artist=True,
                 boxprops=dict(facecolor='lightblue', color='blue'),  # Box color
                 whiskerprops=dict(color='green', linewidth=2),      # Whisker color and line width
                 capprops=dict(color='black', linewidth=2),          # Cap (ends of whiskers) color
                 medianprops=dict(color='red', linewidth=2),         # Median line color
                 flierprops=dict(marker='o', color='purple', markersize=8)) # Outlier color
plt.title('Sales Distribution by Region')
plt.suptitle('')  
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()


plt.figure(figsize=(10, 6))
df.boxplot(column='Profit', by='Region', patch_artist=True)
plt.title('Profit Distribution by Region')
plt.suptitle('')  
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

### Line Plot

```python
plt.figure(figsize=(10, 6)) 

# Plot Sales over time
plt.plot(df['Date'], df['Sales'], label='Sales', marker='o', color='b')

# Adding titles and labels
plt.title('Sales Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Sales', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show grid lines
plt.grid(True)

# Show the legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
