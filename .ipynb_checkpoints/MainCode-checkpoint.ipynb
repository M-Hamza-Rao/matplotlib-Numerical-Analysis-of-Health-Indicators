```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Preaparation
input_file = 'data.csv'
output_file = 'cleaned_data.csv'

with open(input_file, 'r', encoding='utf-8') as f_in, open(output_file, 'w', encoding='utf-8') as f_out:
    for line in f_in:
        # Split by comma to count columns
        columns = line.strip().split(',')
        # Check if line has exactly 34 columns
        if len(columns) == 34:
            f_out.write(line)
```

```python
df = pd.read_csv("cleaned_data.csv")
df.head()
```

```python
def fill_columns(df):
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):  # Check if column is numeric
            df[col].fillna(df[col].mean(), inplace=True)  # Fill NaN with mean
        else:  # Non-numeric columns
            most_frequent = df[col].mode().iloc[0]  # Get the most frequent value
            df[col].fillna(most_frequent, inplace=True)  # Fill NaN with the mode
    return df
```

```python
df.dropna(axis=1, how='all', inplace=True)
df.dropna(how="any", inplace=True)
fill_columns(df)
```

### Task1 

```python
grouped_life_expectancy = df.groupby('ParentLocation')['FactValueNumeric'].mean()
print(grouped_life_expectancy)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(grouped_life_expectancy, bins=10, color='skyblue', edgecolor='black')

# Adding titles and labels
max=math.ceil(grouped_life_expectancy.max())
min=int(grouped_life_expectancy.min())
plt.xticks(np.arange(min, max+1, 2), fontsize=7)
plt.title('Histogram of FactValueNumeric by Region', fontsize=14)
plt.xlabel('Life Expectancy', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Task2

```python
plt.figure(figsize=(4, 3))
x = df['FactValueNumericLow']
y = df['FactValueNumericHigh']
plt.scatter(x, y, color='blue', alpha=0.3, s=5)

#plt.xticks(np.arange(100, 801, 50), fontsize=7)
coefficients = np.polyfit(x, y, 1)  # 1 = linear
slope, intercept = coefficients

trendline = slope * x + intercept

plt.plot(x, trendline, color='red', linewidth=1, label='Trendline')
plt.title('Low vs. High Life Expectancy Scatter Plot', fontsize=7)
plt.xlabel('FactValueNumericLow', fontsize=6)
plt.ylabel('FactValueNumericHigh', fontsize=6)

plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```

### Task3

```python
sales_by_region = df.groupby('Dim1')['FactValueNumeric'].mean()


plt.figure(figsize=(4, 6))
sales_by_region.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])


plt.title('Mean Life Expectancy by Gender', fontsize=12)
plt.xlabel('Life Expectancy', fontsize=12)
plt.ylabel('Genders', fontsize=12)
plt.yticks(np.arange(0, math.ceil(sales_by_region.max()+5), 2), fontsize=7)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### Task4 

##### Box Plot for genders

```python
plt.figure(figsize=(8, 4))
box = df.boxplot(column='FactValueNumeric', by='Dim1', patch_artist=True,
                 boxprops=dict(facecolor='lightblue', color='blue'),  # Box color
                 whiskerprops=dict(color='green', linewidth=2),      # Whisker color and line width
                 capprops=dict(color='black', linewidth=2),          # Cap (ends of whiskers) color
                 medianprops=dict(color='red', linewidth=2),         # Median line color
                 flierprops=dict(marker='o', color='purple', markersize=8)) # Outlier color
plt.yticks(np.arange(0,120, 10), fontsize=7)
plt.title('Box Plot for Gernders')
plt.suptitle('')  
plt.xlabel('Gernders')
plt.ylabel('Life Expectancy')
plt.show()
```
